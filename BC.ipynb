#本文件的目的：先学习state->action
state:状态 action行动
文件：模型学习一系列的点根据当前状态走向(0,0)

```python
import torch
def generate_batch(batch_size=256):
    """
    目的：生成一系列的模仿学习数据集
    :return:state,expert_action
    """
    state=torch.rand(batch_size, 2)*2-1
    goal=torch.zeros_like(state)
    expert_actions=goal-state
    return state,expert_actions
```

```python
import torch.nn as nn
import torch.nn.functional as F
class BCNet(nn.Module):
    def __init__(self):
        super(BCNet, self).__init__()
        self.net=nn.Sequential(
            nn.Linear(2,1024),
            nn.ReLU(),
            nn.Linear(1024,64),
            nn.ReLU(),
            nn.Linear(64,2),
        )
    def forward(self,x):
        return self.net(x)
model=BCNet()
optimizer=torch.optim.Adam(model.parameters())
epoch=500
model.train()
for i in range(epoch):
    state,expert_actions=generate_batch()
    pred_action=model(state)
    loss=F.mse_loss(pred_action,expert_actions)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(loss)
```

```python
state=torch.tensor([-1.2,1.3])
traj=[]
for t in range(300):
    with torch.no_grad():
        action=model(state)
        state=state+0.1*action
        traj.append(state)
print(f'final state:{state}')
print(traj)
```
